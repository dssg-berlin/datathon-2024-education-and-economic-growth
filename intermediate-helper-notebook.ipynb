# Intermediate Level
In this level, you are able to fetch additional metrics from the World Bank Group API (wbgapi) and append it to our dataset

```python
import wbgapi as wb   # I use wb as a namespace in all my work
import pandas as pd   # optional but highly recommended for wbgapi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the main dataset
df = pd.read_csv('dataset.csv')
df.head()
```

```python
# Fetch the new metric
# Set the country and metric
metric_mapping = {'code':'NY.GDP.PCAP.CD', 'name':'GDP per capita (current US$)'}
country_code = 'BRA'

# Use the API to create the sub dataset
new_metric = pd.melt(wb.data.DataFrame(metric_mapping['code'], [country_code], range(1970, 2023), index='time').reset_index(),
                     id_vars=['time'], value_vars=[country_code]).rename(columns={'variable':'Country Code',
                                                                           'time':'year'
                                                                          })
new_metric['year'] = new_metric['year'].str.strip().str[2:]
new_metric['Indicator Code'] = metric_mapping['code']
new_metric['Indicator Name'] = metric_mapping['name']
new_metric['Country Name'] = wb.economy.Series(db=12, id=country_code).iloc[0]
new_metric = new_metric[['Country Name','Country Code','Indicator Name','Indicator Code','year','value']]
new_metric.head()
```

```python
# Append to the main dataset
df = pd.concat([df, new_metric], ignore_index=True)
df
```
